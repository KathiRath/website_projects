## Prepare Notebook
 

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

%matplotlib inline
```

## Generate Data

### 1. Time Variable
 
Let us define a "time" variable $t\in \mathbb{N}$. 

```python
# Number of samples. 
n = 2000
# Generate "time" variable. 
t = np.arange(n)

data_df = pd.DataFrame({'t' : t})
```

### 2. Seasonal Components

Let us define a function to generate seasonal components (Fourier modes).  

```python
# Generate seasonal variables. 
def seasonal(t, amplitude, period):
    """Generate a sinusoidal curve."""
    y1 = amplitude * np.sin((2*np.pi)*t/period) 
    
    return(y1)

# Add two seasonal components. 
data_df ['s1'] = data_df['t'].apply(lambda t : seasonal(t, amplitude=2, period=90))
data_df ['s2'] = data_df['t'].apply(lambda t : seasonal(t, amplitude=1, period=30))

# Define target variable. 
data_df ['y'] = data_df ['s1'] + data_df ['s2'] 
```

Let us plot these components:

```python
plt.rcParams['figure.figsize'] = [12, 8]

fig, ax = plt.subplots()
sns.lineplot(x='t', y='s1', data=data_df, label='s1', ax=ax) 
sns.lineplot(x='t', y='s2', data=data_df, label='s2', ax=ax)
ax.set(title='Seasonal Componets', xlabel='t', ylabel='')
ax.legend(loc='lower left');
```

Now we plot the sum:

```python
fig, ax = plt.subplots()
sns.lineplot(x='t', y='y', data=data_df, ax=ax) 
ax.set(title='Sum Seasonal Componets', xlabel='t', ylabel='');
```

### 3. Trend Component

Next we generate a non-linear trend component:

```python
# Generate trend component. 
def trend(t):
    
    trend_comp = 0.5*(np.power(t, 1/2))
    
    return(trend_comp)

data_df ['y3'] = data_df['t'].apply(trend)
```

Let us see the trend plot:

```python
fig, ax = plt.subplots()
sns.lineplot(x='t', y='y3', data=data_df, ax=ax) 
ax.set(title='Trend Component', xlabel='t', ylabel='');

# Add trend to target variable. 
data_df ['y'] = data_df ['y'] + data_df ['y3']
```

### 4. Add Gaussian Noise

```python
# Set noise standard deviation. 
sigma_n = 0.3

data_df['epsilon'] = np.random.normal(loc=0, scale=sigma_n, size=n)
# Add noise to target variable. 
data_df ['y'] = data_df ['y'] + data_df ['epsilon']
```

```python
fig, ax = plt.subplots()
sns.lineplot(x='t', y='y', data=data_df, ax=ax)
ax.set(title='Sample Data', xlabel='t', ylabel='');
```

## Define Kernel Structure 

We now construct the predicton kernel:

```python
from sklearn.gaussian_process.kernels \
    import RBF, WhiteKernel, ExpSineSquared, ConstantKernel, RationalQuadratic


k1 = ConstantKernel(constant_value=2, constant_value_bounds=(1e-2, 1e5)) * RBF(length_scale=10, length_scale_bounds=(1e-4, 1e4)) 

k2 = ConstantKernel(constant_value=2) * ExpSineSquared(length_scale=1.0, periodicity=90, periodicity_bounds=(80, 100))

k3 = ConstantKernel(constant_value=1) * ExpSineSquared(length_scale=1.0, periodicity=30, periodicity_bounds=(25, 35))

# Add all individual kernel components. 
kernel  = k1 + k2 + k3
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
 
# Define GaussianProcessRegressor object. 
gp = GaussianProcessRegressor(
    kernel=kernel, 
    n_restarts_optimizer=3, 
    normalize_y=True,
    alpha=0.3
)
```

```python
X = data_df['t'].values.reshape(n ,1)
y = data_df['y'].values.reshape(n ,1)

n_train = round(0.75*n)

X_train = X[:n_train]
y_train = y[:n_train]

X_test = X[n_train:]
y_test = y[n_train:]
```

```python
gp.fit(X_train, y_train)
```

```python
y_pred_train = gp.predict(X_train) 
y_pred_test = gp.predict(X_test) 

y_pred, y_std = gp.predict(X, return_std=True)

data_df['y_pred'] = y_pred
data_df['y_std'] = y_std
data_df['y_pred_lwr'] = data_df['y_pred'] - data_df['y_std']
data_df['y_pred_upr'] = data_df['y_pred'] + data_df['y_std']
```

```python
fig, ax = plt.subplots(figsize=(15, 8))

sns.lineplot(x='t', y='y', data=data_df, label = 'y', ax=ax)
sns.lineplot(x='t', y='y_pred', data=data_df, label='y_pred', color='green', ax=ax)

ax.fill_between(
    x=data_df['t'], 
    y1=data_df['y_pred_lwr'], 
    y2=data_df['y_pred_upr'], 
    color='green', 
    alpha=0.3
)

ax.axvline(n_train, color='red', linestyle='--')
ax.set(title='Prediction', xlabel='t', ylabel='')
ax.legend(loc='upper left');
```
